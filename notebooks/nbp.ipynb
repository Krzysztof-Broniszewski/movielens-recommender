```python
import requests
import pandas as pd

url = "https://api.nbp.pl/api/exchangerates/tables/A/"

response = requests.get(url, timeout=10)
response.raise_for_status()

data = response.json()[0]

df_nbp = pd.DataFrame(data["rates"])
df_nbp["effectiveDate"] = data["effectiveDate"]
df_nbp["tableNo"] = data["no"]
df_nbp = df_nbp[["effectiveDate", "tableNo", "currency", "code", "mid"]]

df_nbp.head(100)
```
